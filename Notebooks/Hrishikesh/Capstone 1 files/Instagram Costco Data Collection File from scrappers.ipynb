```python
import pandas as pd
import os
```

```python
# File paths
post_urls_file_path = "instagram_post_urls_updated_1877.txt"
result_file_path = "result.csv"

# Load the Instagram post URLs from the text file
with open(post_urls_file_path, "r") as file:
    unique_post_urls = set(line.strip() for line in file if line.strip())

# Load the results CSV file
df_new = pd.read_csv(result_file_path)

# Extract unique post URLs from the 'query' column in the results CSV file
scraped_post_urls_new = set(df_new["query"].dropna().unique())

# Count how many unique post URLs were scraped
scraped_count_new = len(unique_post_urls.intersection(scraped_post_urls_new))

scraped_count_new
```

```python
# Identify processed and unprocessed URLs
processed_urls = unique_post_urls.intersection(scraped_post_urls_new)
unprocessed_urls = unique_post_urls - processed_urls

# Save processed URLs to a file
with open("processed_urls.txt", "w") as file:
    for url in processed_urls:
        file.write(url + "\n")

# Save unprocessed URLs to a file
with open("unprocessed_urls.txt", "w") as file:
    for url in unprocessed_urls:
        file.write(url + "\n")

print(f"Processed URLs: {len(processed_urls)}")
print(f"Unprocessed URLs: {len(unprocessed_urls)}")
print("Files 'processed_urls.txt' and 'unprocessed_urls.txt' have been created.")
```
